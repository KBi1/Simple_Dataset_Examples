```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Defining a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Favorite Fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Banana', 'Orange', 'Apple']
}
```

```python
# Creating a DataFrame from the dataset
df = pd.DataFrame(data)
```

```python
# Display the dataset
print("Dataset Info:")
print(df)
```

```python
# Counting the occurrences of each fruit in the 'Favorite Fruit' column
fruit_counts = df['Favorite Fruit'].value_counts()
```

```python
# Plotting a bar chart using the counts
plt.bar(fruit_counts.index, fruit_counts.values, color=['red', 'yellow', 'orange'])

# Adding labels to the x and y axes
plt.xlabel('Favorite Fruit')
plt.ylabel('Number of People')

# Adding a title to the plot
plt.title('Favorite Fruits Distribution')


# Displaying the plot
plt.show()
```
