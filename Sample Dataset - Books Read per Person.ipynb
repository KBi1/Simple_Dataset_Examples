```python
# Importing the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Books Read': [3, 1, 5, 2, 4, 2, 6]
}
```

```python
# Create a DataFrame
df_books = pd.DataFrame(data)
```

```python
# Display the dataset
print("Dataset:")
print(df_books)
```

```python
# Plotting a bar chart for books read
plt.bar(df_books['Name'], df_books['Books Read'], color='#9cc1b5') # You can use a variety of color names like 'red',etc., or you can use hex color codes

# Set the label for the x-axis
plt.xlabel('Name')

# Set the label for the y-axis
plt.ylabel('Number of Books Read')

# Set the title of the plot
plt.title('Books Read per Person')

# Display the plot
plt.show()
```
